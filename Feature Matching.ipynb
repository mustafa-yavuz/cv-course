```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
reeses = cv2.imread('reeses_puffs.png',0)     
display(reeses)
```

```python
cereals = cv2.imread('many_cereals.jpg',0) 
display(cereals)
```

```python
# Brute Force Detection with ORB Descriptors

# Initiate ORB detector
orb = cv2.ORB_create()

# find the keypoints and descriptors with ORB
kp1, des1 = orb.detectAndCompute(reeses,None) # None -> No mask
kp2, des2 = orb.detectAndCompute(cereals,None) # None -> No mask

# create BFMatcher object
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

# Match descriptors.
matches = bf.match(des1,des2) 

# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance) # less distance is a better match

# Draw first 25 matches.
reeses_matches = cv2.drawMatches(reeses,kp1,cereals,kp2,matches[:25],None,flags=2) 
```

```python
display(reeses_matches) # not really good matches
```

```python
# Brute-Force Matching with SIFT Descriptors and Ratio Test
# Much better when sizes of inpt and target images are different

# Create SIFT Object
sift = cv2.xfeatures2d.SIFT_create()

# find the keypoints and descriptors with SIFT
kp1, des1 = sift.detectAndCompute(reeses,None)
kp2, des2 = sift.detectAndCompute(cereals,None)

# BFMatcher with default params
bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2, k=2)

# Apply ratio test
good_matches = []
for match1,match2 in matches:
    if match1.distance < 0.75*match2.distance:
        good_matches.append([match1])

# cv2.drawMatchesKnn expects list of lists as matches.
sift_matches = cv2.drawMatchesKnn(reeses,kp1,cereals,kp2,good_matches,None,flags=2)
```

```python
display(sift_matches)
```

```python
# FLANN based Matcher
# Finds all matches not best. so its faster but not necessarily the besy

# Initiate SIFT detector
sift = cv2.xfeatures2d.SIFT_create()

# find the keypoints and descriptors with SIFT
kp1, des1 = sift.detectAndCompute(reeses,None)
kp2, des2 = sift.detectAndCompute(cereals,None)

# FLANN parameters
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks=50)  

flann = cv2.FlannBasedMatcher(index_params,search_params)

matches = flann.knnMatch(des1,des2,k=2)

good_mathces = []

# ratio test
for i,(match1,match2) in enumerate(matches):
    if match1.distance < 0.7*match2.distance:
        
        good_matches.append([match1])


flann_matches = cv2.drawMatchesKnn(reeses,kp1,cereals,kp2,good_matches,None,flags=0) # flags 0 show all points flags 2 dont

display(flann_matches) # finds all matches but only draws what it thinks are the best matches
```

```python
# Need to draw only best matches, so creating a mask

matchesMask = [[0,0] for i in range(len(matches))]

# ratio test
for i,(match1,match2) in enumerate(matches):
    if match1.distance < 0.7*match2.distance: 
        matchesMask[i]=[1,0]  # labels the lines(best) as 1

draw_params = dict(matchColor = (0,255,0),
                   singlePointColor = (255,0,0),
                   matchesMask = matchesMask,
                   flags = 0)

flann_matches = cv2.drawMatchesKnn(reeses,kp1,cereals,kp2,matches,None,**draw_params)

display(flann_matches)
```
