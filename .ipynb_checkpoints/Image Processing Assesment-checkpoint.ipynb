```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Function to display image
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
# Open and display the image
img = cv2.imread('giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display_img(img)
```

```python
# Applying a binary threshold onto the grayscale image (not all channels)
img = cv2.imread('giraffes.jpg',0) 
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
display_img(thresh1,cmap='gray')
```

```python
# Converting colorspace to HSV
img = cv2.imread('giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
display_img(img)
```

```python
# 4x4 kernel
kernel = np.ones(shape=(4,4),dtype=np.float32)/10
```

```python
kernel
```

```python
img = cv2.imread('giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# 2D Conv to blur the image
dst = cv2.filter2D(img,-1,kernel)
display_img(dst)
```

```python
# Creating a horizontal sobel filter with a kernel size of 5 to the grayscale image
img = cv2.imread('giraffes.jpg',0)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)

# gradient filtered version (works as edge detection)
display_img(sobelx,cmap='gray')
```

```python
# Plotting color histograms for both 3 channels

img =cv2.imread('giraffes.jpg')

color = ('b','g','r')

for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
    
plt.title('Giraffes Histograms')
plt.show()
```
